### Add markers

https://matplotlib.org/stable/api/markers_api.html

## Legend

```python
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.xlabel("x", fontsize=12)
plt.ylabel("sin/cos", fontsize=12)
plt.title("sin/cos curves", fontsize=18)
plt.legend();
```

### set `loc, ncol, fontsize`

```python
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.xlabel("x", fontsize=12)
plt.ylabel("sin/cos", fontsize=12)
plt.title("sin/cos curves", fontsize=18)
plt.legend(loc="upper center", ncol=2, fontsize=14);
```

```python
plt.plot(x, np.sin(x), label="sin")
plt.plot(x, np.cos(x), label="cos")
plt.xlabel("x", fontsize=12)
plt.ylabel("sin/cos", fontsize=12)
plt.title("sin/cos curves", y = -0.3, fontsize=18)
plt.legend(loc="upper center", ncol=2, fontsize=14);
```

```python
df = pd.read_csv("iris.csv")
df.head(5)
```

```python
plt.scatter(df["sepal_length"], df["sepal_width"],c=df["species"].astype("category").cat.codes, alpha=0.9)
plt.xlabel("sepal length")
plt.ylabel("sepal width");
```

Complete the above plot by adding the following:
- legend for each species
- title in the bottom


```python
species = df["species"].unique()

# Plot each species separately
for i, species_name in enumerate(species):
    species_data = df[df["species"] == species_name]
    plt.scatter(species_data["sepal_length"], species_data["sepal_width"], 
                label=species_name, alpha=0.9)

# Label the axes and title
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset")

# Add legend
plt.legend(title="Species");
```

## Bar plot

Bar plots represent categorical data with rectangular bars where the height (or length for horizontal bars) is proportional to the values they represent.

They are used to compare quantities across different categories.

`plt.bar(), plt.barh()`

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]
```

```python
plt.bar(categories, values);
```

```python
plt.barh(categories, values);
```

```python
df.head(5)
```

```python
avg_iris = df.groupby("species")["sepal_length"].mean()
avg_iris
```

```python
plt.barh(avg_iris.index, avg_iris);
```

```python
avg_iris = df.groupby("species").mean()
avg_iris
```

```python
width=0.2
multiplier=0
x = np.arange(3)
for i in range(avg_iris.shape[1]):
    offset = width*multiplier
    plt.bar(x + offset, avg_iris.iloc[:,i], width, label=avg_iris.columns[i])
    multiplier += 1
plt.legend()
plt.xticks(x + width, avg_iris.index);
```

## Histograms

Histograms are used to display the distribution of a dataset. They group data into bins and count how many data points fall into each bin.

`plt.hist(x, bins=None, density=False, histtype='bar')`

```python
x = np.random.normal(size=1000)
```

```python
plt.hist(x);
```

```python
plt.hist(x, bins=30);
```

What is the best value for bins?

```python
plt.hist(x, bins=30, density=True); # y-scale become 1
```

```python
plt.hist(x, bins=30, density=True, alpha=0.8);
```

We can also specify `histtype` which has four options `{'bar', 'barstacked', 'step', 'stepfilled'}`. The default is `'bar'`.

```python
plt.hist(x, bins=30, density=True, alpha=0.5, histtype="stepfilled", edgecolor="black");
```

```python
plt.hist(x, bins=30, density=True, cumulative=True, alpha=0.3, histtype="bar", edgecolor="black");
```

```python
x1 = np.random.normal(0,1,size=1000)
x2 = np.random.normal(0,5, size=1000)
x3 = np.random.normal(-5, 1, size=1000)
```

```python
for i in [x1, x2, x3]:
    plt.hist(i, bins=30, density=True, alpha=0.5);
```

```python
X = np.vstack([x1, x2, x3]).T
X.shape
```

```python
plt.hist(X, bins=100, density=True, alpha=0.5, histtype="barstacked");
```

## Boxplot

Boxplots (or whisker plots) are used to display the distribution of data based on five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.

They provide a visual way to identify outliers, spread, and skewness.

`plt.boxplot(data, vert=True, patch_artist=False)`

```python
plt.boxplot(x);
```

## Contour plot

Contour lines represent different levels of a third variable (Z) over a grid defined by two variables (X and Y).

`plt.contourf()`

```python
x = y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2
plt.contourf(X, Y, Z, cmap=plt.cm.RdPu)
plt.axis('square');
```

# Visualize uncertainty

`plt.errorbar(x, y, yerr=None, xerr=None,...)`

```python
x = np.linspace(0, 10, 20)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(20)
```

```python
plt.errorbar(x, y, yerr=dy, fmt=".");
```

```python
plt.errorbar(x, y, yerr=dy, fmt="o", color="black", ecolor="lightgray");
```

```python
dx = 0.2
```

```python
plt.errorbar(x, y, yerr=dy, xerr=dx, fmt="o", color="black", ecolor="darkgray");
```

### `plt.fill_between` 

create a shaded region between two sets of y-values.

`plt.fill_between(x, y1, y2)`

```python
x = np.linspace(1,10,10)
y_mean = np.linspace(1, 10, 10)

y_upper = y_mean + 1.5
y_lower = y_mean - 1.5
```

```python
plt.plot(x, y_mean, color="blue")
plt.fill_between(x, y_upper, y_lower, color="lightblue");
```
